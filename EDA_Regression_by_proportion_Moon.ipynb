```python
df = pd.read_csv("Life_Expectancy_preprocessed.csv")
df = df.iloc[:, 1:]

df
```

```python
df.describe()
```

```python
# corelation
import matplotlib.pyplot as plt
import seaborn as sns

colormap = plt.cm.PuBu
plt.figure(figsize=(12, 10))
plt.title("Pearson Correlation of Features", y = 1.05, size = 15)
sns.heatmap(df.iloc[:, 1:].corr(), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 8})
```

```python
correlation_matrix = df.iloc[:, 1:].corr()

# Extract values whose correlation exceeds threshold(ex: 0.5)
target_column = 'Life expectancy '
threshold = 0.4
columns_above_threshold = correlation_matrix.index[abs(correlation_matrix[target_column]) > threshold]

# print result
print("\nVariables whose correlation with the target variable exceeds {}:".format(threshold))
print(columns_above_threshold)
```

Regression

```python
columns_above_threshold = list(columns_above_threshold)
columns_above_threshold.remove('Life expectancy ')
# remove target value from list
```

```python
X = df[columns_above_threshold]
y = df.iloc[:, -1]
y = y.values.reshape(-1, 1)
X.head()
# chosen variables only
# divide dataset into X(independent) and y(target)
```

```python
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


# Save model performance for different ratios
train_ratios = np.arange(0.1, 1.0, 0.1)  # ratios (from 0.1 to 0.9, 0.1 interval)
r2_scores = []
rmse_list = []

for train_ratio in train_ratios:
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_ratio, random_state=42)
    
    # train model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # evaluate model
    y_pred = model.predict(X_test)
    score = r2_score(y_test, y_pred)
    r2_scores.append(score)    # save score in the list
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    rmse_list.append(rmse)     # save rmse in the list

```

```python
# draw a graph
plt.figure(figsize=(10, 6))
plt.plot(train_ratios, r2_scores, marker='o', linestyle='-')
plt.title('Model Performance vs. Test Set Ratio')
plt.xlabel('Train Set Ratio')
plt.ylabel('R2 score')
plt.grid(True)
plt.show()
```

```python
# graph
plt.figure(figsize=(10, 6))
plt.plot(train_ratios, rmse_list, marker='o', linestyle='-')
plt.title('Loss value vs. Train Set Ratio')
plt.xlabel('Train Set Ratio')
plt.ylabel('RMSE')
plt.grid(True)
plt.show()
```

The model performs best at a test set ratio of 0.4 to 0.6.
I expected more than 0.9 but not as high as I expected.
And when it comes to error, it grdually increases slightly.

This is because underfitting occurs when there are few training sets, and overfitting occurs when there are too many training sets, making it difficult to predict test sets.

Generally, the proportion of training sets is often 0.6 or 0.7, but in this case 0.5 looks best when comparing performance and error.
